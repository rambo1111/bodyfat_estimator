```python
# Extract the columns you want for the scatter plot matrix
columns_for_scatter = ['Density', 'BodyFat', 'Age', 'Weight', 'Height', 'Neck', 'Chest',
                        'Abdomen', 'Hip', 'Thigh', 'Knee', 'Ankle', 'Biceps', 'Forearm',
                        'Wrist']

# Create a new DataFrame with only the selected columns
scatter_data = df[columns_for_scatter]

# Create a scatter plot matrix
sns.set(style="ticks")
sns.pairplot(scatter_data)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Extract features (X) and target (y)
X = df.drop('BodyFat', axis=1)  # Features excluding BodyFat
y = df['BodyFat']

# # Split the data into training and testing sets
# X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=69)
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
# Evaluate the model on the test set
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
print(f'Mean Squared Error on Test Set: {mse}')
```

```python
import pickle

# Save the model to a file using pickle
with open('bodyfat.pkl', 'wb') as file:
    pickle.dump(model, file)
```
